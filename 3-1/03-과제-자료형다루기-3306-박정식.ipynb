```python
a = 10
b = -2.5
c = 1 + 2j
d = 0xDA # 218

print(a, type(a))
print(b, type(b))
print(c, c.real , c.imag, c.conjugate(), type(c))
print(d, type(d))

print(a + b, type(a + b))
print(c + d, type(c + d))
```

a~d의 각 값과 타입을 출력했을 때 위에서 살펴본 정수형과 실수형, 복소수형으로 출력되는 것을 확인할 수 있습니다. 이 때 정수+실수는 실수형으로, 정수+복소수는 복소수의 형태로 변한다는 것을 함께 확인하시기 바랍니다.
