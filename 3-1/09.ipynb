## Pandas
1) 데이터 조작 및 분석을 위한 파이썬 프로그래밍 언어 용으로 작성된 소프트웨어 라이브러리<br>
2) 특히 숫자 테이블과 시계열을 조작하기 위한 데이터 구조와 연산을 제공함<br>
3) 판다스는 파이썬에서 자료 정제 및 분석을 빠르고 쉽게 할 수 있는 자료 구조와 관리 도구들을 가지고 있음<br>
###■판다스는 
1) 수치 해석 도구들인 NumPy, SciPy 및 <br>
2) 자료 분석 라이브러리인 statsmodels, scikit-learn과 <br>
3) 시각화 도구인 matplotlib들과 함께 사용된다. 
####■ 판다스는 NumPy에서 사용되는 배열 기반 연산과 for 문 없이 계산하는 벡터화 계산을 기반으로 한다. 
####■ 넘파이와 판다스의 큰 차이점: 넘파이는 같은 형에 대한 배열을 주로 다루는 반면 판다스는 이질적인 자료형을 다루는 것에 더 적합하다는 것.
