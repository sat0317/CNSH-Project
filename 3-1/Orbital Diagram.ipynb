```python
atomicList=('H', 'He', 'Li', 'Be', 'B', 'C', 'N', 'O', 'F', 'Ne', 'Na', 'Mg', 'Al', 'Si', 'P', 'S', 'Cl', 'Ar', 'K', 'Ca', 'Sc', 'Ti', 'V', 'Cr', 'Mn', 'Fe', 'Co', 'Ni', 'Cu', 'Zn', 'Ga', 'Ge', 'As', 'Se', 'Br', 'Kr', 'Rb', 'Sr', 'Y', 'Zr', 'Nb', 'Mo', 'Tc', 'Ru', 'Rh', 'Pd', 'Ag', 'Cd', 'In', 'Sn', 'Sb', 'Te', 'I', 'Xe', 'Cs', 'Ba', 'La', 'Ce', 'Pr', 'Nd', 'Pm', 'Sm', 'Eu', 'Gd', 'Tb', 'Dy', 'Ho', 'Er', 'Tm', 'Yb', 'Lu', 'Hf', 'Ta', 'W', 'Re', 'Os', 'Ir', 'Pt', 'Au', 'Hg', 'Ti', 'Pb', 'Bi', 'Po', 'At', 'Rn', 'Fr', 'Ra', 'Ac', 'Th', 'Pa', 'U', 'Np', 'Pu', 'Am', 'Cm', 'Bk', 'Cf', 'Es', 'Fm', 'Md', 'No', 'Lr', 'Rf', 'Db', 'Sg', 'Bh', 'Hs', 'Mt', 'Ds', 'Rg', 'Cn', 'Nh', 'Fl', 'Mc', 'Lv', 'Ts', 'Og')
orbitalName = ['s', 'p', 'd', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o']
exceptList = {
    'Cr': "1s2 2s2 2p6 3s2 3p6 4s1 3d5",
    'Cu': "1s2 2s2 2p6 3s2 3p6 4s1 3d10",
    'Mo': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s1 4d5",
    'Ru': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s1 4d7",
    'Rh': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s1 4d8",
    'Pd': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 4d10",
    'Ag': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s1 4d10",
    'La': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 5d1",
    'Ce': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f1 5d1",
    'Gd': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f7 5d1",
    'Pt': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s1 4f14 5d9",
    'Au': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s1 4f14 5d10",
    'Ac': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f14 5d10 6p6 7s2 6d1",
    'Th': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f14 5d10 6p6 7s2 6d2",
    'Pa': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f14 5d10 6p6 7s2 5f2 6d1",
    'U':  "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f14 5d10 6p6 7s2 5f3 6d1",
    'Np': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f14 5d10 6p6 7s2 5f4 6d1",
    'Cm': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f14 5d10 6p6 7s2 5f7 6d1",
    'Lr': "1s2 2s2 2p6 3s2 3p6 4s2 3d10 4p6 5s2 4d10 5p6 6s2 4f14 5d10 6p6 7s2 5f14 7p1"
}

atomicNum = {}
for i in range(1, len(atomicList)+1):
    print(i, atomicList[i-1], end='\t')
    atomicNum[atomicList[i-1]]=i
```

```python
target=input("원소를 입력하세요: ")
```

```python
orbital=""

if target not in atomicList:
    print("잘못된 원소입니다.")
    exit()
elif target not in exceptList.keys():
    k=atomicNum[target]
    breakPoint=False
    for ns in range(1, 100):
        if breakPoint:
            break
        for n in range(1, ns+1, 1):
            s=ns-n
            if k==0:
                break
            if n<=s or s<0:
                continue
           # print(n, s)
            if (s)*4+2>k:
                orbital+=' '+str(n)+orbitalName[s]+str(k)
                breakPoint=True
                break
            else:
                orbital+=' '+str(n)+orbitalName[s]+str((s)*4+2)
                k-=s*4+2

else:
    orbital = exceptList[target]
orbital = orbital.strip()
print(target, ':', orbital)
```

```python
splitedOrbital = orbital.split()
linkedOrbital = []
maxPos=2
for i in range(len(splitedOrbital)):
    linkedOrbital.append((int(splitedOrbital[i][0]), 
                          orbitalName.index(splitedOrbital[i][1]), 
                          int(splitedOrbital[i][2:])))
    if maxPos<linkedOrbital[i][1]:
        maxPos=linkedOrbital[i][1]
        
maxNorbital=len(splitedOrbital)+1

plt.style.use('default')
plt.rcParams['figure.figsize'] = (maxPos+maxPos*(maxPos+1)+2.3, (maxNorbital))
fig, ax=plt.subplots()

ax.set_xlim(0, maxPos+maxPos*(maxPos+1)+2.3)
ax.set_ylim(0, maxNorbital)
plt.title('Orbital Diagram of ' + target, size=28, font='serif', pad = 15)
for i in range(len(linkedOrbital)):
    maxRight=0
    for j in range(1, (linkedOrbital[i][1]*2+1)*2+1):
        if j>1+linkedOrbital[i][1]*2:
            if j<=linkedOrbital[i][2]:
                ax.add_patch(
                    patches.Arrow(
                        x=(linkedOrbital[i][1]**2)+j-(1+linkedOrbital[i][1]*2)+0.3, 
                        y=i+1+0.6,
                        dx=0, dy=-0.6,
                        width=0.2,
                        color='black'
                    )
                )
        else:
            if j<=linkedOrbital[i][2]:
                ax.add_patch(
                    patches.Arrow(
                        x=(linkedOrbital[i][1]**2)+j, y=i+1,
                        dx=0, dy=0.6,
                        width=0.2,
                        color='black'
                    )
                )
            ax.add_patch(
                patches.Rectangle(
                    ((linkedOrbital[i][1]**2)+j-0.2, i+1-0.05),
                    0.7, 0.05,
                    fill = True,
                    facecolor = 'black',
                ))
            maxRight=max(maxRight, linkedOrbital[i][1]**2+j)
    plt.text(maxRight+0.6, i+1-0.1, 
            str(linkedOrbital[i][0])+orbitalName[linkedOrbital[i][1]], 
            fontsize=13, font='serif')


plt.show()
```
